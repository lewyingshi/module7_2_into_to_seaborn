### How to make a Seaborn swarmplot?

Seaborn's `swarmplot()` function a.k.a the "beeswarm" plot style, is similar to `stripplot()`, but the points are adjusted so that they don't overlap. This gives a much better representation of the distribution of values, but it does not scale well to large number of observations.

- can be used to complement other plots (e.g boxplots or violin plots)
- styling
    - linewidth
    - edgecolor

```python
cars = sns.load_dataset('mpg').dropna()
cars.head()
```

```python
sns.get_dataset_names()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = (cars >> filter_by(X.cylinders.isin([4,6,8])))
cars.cylinders.value_counts()
```

```python
sns.swarmplot(x = cars.horsepower);
```

hue parameter determines which column in the data frame should be used for colour encoding.

```python
sns.swarmplot(data=cars, 
              x='origin',
              y='horsepower', 
              hue='cylinders');
```

use dodge to split groups to observe each swarm individually

```python
sns.swarmplot(data=cars, 
              x='origin', 
              y='horsepower', 
              hue='cylinders',
              dodge = True);
```

#### Overlay swarm plot with boxplots or violinplots

```python
# with box plots
usa = (cars >> filter_by(X.origin == 'usa'))

sns.boxplot(x=usa.cylinders, y=usa.horsepower,
            whis=np.inf) 

sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
              color='black', alpha=0.5, size=4);
```

```python
# with violin plots
sns.violinplot(x=usa.cylinders, y=usa.horsepower,
               scale='width', inner=None, color='xkcd:forest green')

sns.swarmplot(x=usa.cylinders, y=usa.horsepower, 
              color='white', size=3);
```

#### Styling

```python
sns.swarmplot(data=cars, x='horsepower', linewidth=1, edgecolor='black');
```

swarmplot inherits from the matplotlib scatter properties

```python
plt.figure(figsize=(8,4))
sns.swarmplot(data=cars, x='horsepower', marker='*')
```

### How to make a Seaborn lineplot?

Seaborn's `lineplot()` is used to illustrate a continuous relationship between two values. Time is most of the time one of the values and plotted at the `x` axis. 

By using `lineplot()`, you can easily find patterns, trends or even seasonality in the data.

```python
# parking data in Birmingham, UK
df = pd.read_csv("./data/dataset.csv", parse_dates=[3])
df.head()
```

```python
df.rename(columns={'SystemCodeNumber': 'Location', 'LastUpdated': 'Timestamp'}, inplace=True)
df['Day'] = df.Timestamp.dt.date
df['Month'] = df.Timestamp.dt.month
df['Hour'] = df.Timestamp.dt.hour
```

```python
park = df[df.Location.isin(['Broad Street', 'NIA South'])]
park.head()
```

```python
park.info()
```

```python
sns.set_style('dark')
```

lineplot that represents the mean value of cars and garages of each day of the year.

```python
sns.lineplot(data=park, x='Day', y='Occupancy');
```

look at occupancy throughout the day

```python
sns.lineplot(data=park, x='Hour', y='Occupancy');
```

```python
sns.lineplot(data = park, x = 'Hour', y = 'Occupancy', ci=None);
```

```python
sns.lineplot(data = park, x = 'Hour', y = 'Occupancy', ci=68);
# you can change how big the confidence interval should be with 
# ci, default is ci = 95
```

```python
sns.lineplot(data = park, x = 'Hour', y = 'Occupancy', estimator='mean');
# accessing estimator argument to look at other aggregate functions
```

```python
sns.lineplot(data = park, x = 'Hour', y = 'Occupancy', estimator = 'sum');
```

```python
sns.lineplot(data = park, x = 'Hour', y = 'Occupancy', estimator = 'std');
```

### Visual Semantics

#### Hue

```python
# additional info to show using visual semantics
sns.lineplot(data = park, x = 'Day', y = 'Occupancy', hue = 'Location');
```

```python
sns.lineplot(data=park, x='Day', y='Occupancy', hue='Location', palette = ['gray', 'xkcd:brick red']);
```

#### Style

```python
sns.lineplot(data = park, x = 'Day', y = 'Occupancy', style = 'Location');
```

```python
sns.lineplot(data = park, x = 'Day', y = 'Occupancy', hue = "Location", style = 'Location');
```

#### Size

```python
sns.lineplot(x='Hour', y='Occupancy', data=park, size='Location');
```

```python
sns.lineplot(x='Hour', y='Occupancy', data=park, ci=None,
             size='Month');
```

```python
sns.lineplot(x='Day', y='Occupancy', data=park, ci=None, 
             lw=4,
             color='#aa00aa',
             alpha=0.5
            );
# adjust transparency with alpha
```
