### Seaborn: swarmplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

filter down to only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
# build swarm
sns.swarmplot(x=cars.horsepower);
```

```python
sns.swarmplot(data=cars, x='origin', y='horsepower');
```

```python
sns.swarmplot(data=cars, x='origin', y='horsepower', hue='cylinders',
              dodge=True);
# dodge is defaulted to False with swarmplots
```

### Swarmplot Overlay

```python
usa = cars[cars.origin == 'usa']
```

#### With Boxplot

```python
sns.boxplot(x=usa.cylinders, y=usa.horsepower,
            whis=np.inf)

sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
              color='black', alpha=0.5, size=4);
```

#### With Violin plot

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower,
               scale='width', inner=None, color='xkcd:forest green')

sns.swarmplot(x=usa.cylinders, y=usa.horsepower, 
              color='white', size=3);
```

### Styling

#### linewidth, edgecolor

```python
sns.swarmplot(data=cars, x='horsepower', linewidth=1, edgecolor='black');
```

### Matplotlib Scatter Properties: marker

swarmplot inherits from the matplotlib scatter properties

```python
plt.figure(figsize=(8,4))
sns.swarmplot(data=cars, x='horsepower', marker='*')
```
